# World Bank

## Indicators and Countries

##### Main categories
/worldBank/categories
https://api.tradingeconomics.com/worldBank/categories?c=guest:guest

```python
te.getWBCategories(output_type='df')
```

##### Filtering by the main categories
/worldBank/category/{category}
https://api.tradingeconomics.com/worldBank/category/Education?c=guest:guest

```python
te.getWBCategories(category='education',output_type='df')
```

##### Detailed information about specific indicator for all countries using a series code from the previous method
/worldBank/indicator?c=guest:guest&s={series}
https://api.tradingeconomics.com/worldBank/indicator?c=guest:guest&s=fr.inr.rinr

```python
te.getWBIndicator(series_code = 'SE.ADT.1524.LT.FE.ZS', output_type = 'df')
```

##### List of indicators available for a specific country
/worldBank/country/{country}
https://api.tradingeconomics.com/worldBank/country/portugal?c=guest:guest

```python
te.getWBCountry(country = 'portugal', output_type = 'df')
```

##### Detailed information just for a specific indicator and country by using series code or url
/worldBank/indicator?c=guest:guest&s={series_code}
https://api.tradingeconomics.com/worldBank/indicator?c=guest:guest&s=usa.fr.inr.rinr

```python
te.getWBIndicator(series_code = 'usa.fr.inr.rinr', output_type = 'df')
```

/worldBank/indicator?c=guest:guest&url={url1}
